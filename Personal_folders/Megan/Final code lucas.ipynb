```python
#er zijn 2 colommen die heel veel NaN weergeven, die ga ik eruit gooien omdat dit niet bruikbaar is.
df = df.drop(['Rape\n(revised \ndefinition3)', 'Rape\n(revised \ndefinition) \nrate3'], axis = 1)
```

```python
df.head()
df = df.dropna()
df
```

```python
df = df.rename(columns = {'Population1' : 'Population', 'Violent\ncrime2' : 'Violent\ncrime'})
df
```

```python
df.plot(x='Population', y='Burglary \nrate ', kind ='line')
plt.xlabel('Population')
plt.ylabel('Burglary \nrate ')
plt.grid()
plt.tight_layout()
plt.title('Correlatie inbraak en populatie')
plt.show()
```

```python
#voorbeeld van veranderende layout
df.plot(x='Population', y='Burglary \nrate ', kind ='scatter',color ='red')
plt.xlabel('Population')
plt.ylabel('Burglary \nrate ')
plt.grid(ms=2)
plt.tight_layout()
plt.title('Correlatie inbraak en populatie')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,14)})
sns.heatmap(df.corr(),annot=True, annot_kws={"size": 8},linewidths=0.1, linecolor='black')
plt.style.use('default')
plt.title('Seaborn Heatmap')
#Alles met hele donkere kleuren is negatieve correlatie
```
