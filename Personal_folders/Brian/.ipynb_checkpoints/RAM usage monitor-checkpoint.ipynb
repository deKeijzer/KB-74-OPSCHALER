```python
%matplotlib inline
import time
import pylab as pl
from IPython import display
values = []
```

```python
for i in range(1000):
    values.append(psutil.virtual_memory()[2])
    pl.plot(values)
    display.clear_output(wait=True)
    display.display(pl.gcf())
    time.sleep(1)
```
