# Reading in the data
The data consists of multiple files with a structured file name.  
Therefor glob is used to get all the file paths, afterwhich Pandas is used to merge the DataFrames.  

```python
import glob
import pandas as pd

dir = '//datc//opschaler//weather_data//knmi_10_min_raw_data//'
pattern = 'keijzer_hhs_****.LST'

files = glob.glob(dir+pattern) # Get a list of all the files
```

```python
files
```

## Extracting file labels from the filepaths

```python
# Get the file labels (tot, toa, tor, tow)

file_labels = []

for file in files:
    file_labels.append([file[-12:-9],file[-8:-4]])

file_labels
```

```python
file_labels[1][0]
```

## Creating a list of all DataFrames

```python
toa_names = ['stn', 'lcd', 'datum', 'tijd', 'WW', 'VV', 'P']
toc_names = ['stn', 'lcd', 'datum', 'tijd', 'N']
tor_names = ['stn', 'lcd', 'datum', 'tijd', 'RG', 'DR']
tos_names = ['stn', 'lcd', 'datum', 'tijd', 'Q', 'SQ']
tot_names = ['stn', 'lcd', 'datum', 'tijd', 'T10', 'TD', 'T', 'U']
tow_names = ['stn', 'lcd', 'datum', 'tijd', 'DD', 'FF', 'FX']

dfs = []

for i,file in enumerate(files):
    if file_labels[i][0] == 'toa':
        column_names = toa_names
    elif file_labels[i][0] == 'toc':
        column_names = toc_names
    elif file_labels[i][0] == 'tor':
        column_names = tor_names
    elif file_labels[i][0] == 'tos':
        column_names = tos_names
    elif file_labels[i][0] == 'tot':
        column_names = tot_names
    elif file_labels[i][0] == 'tow':
        column_names = tow_names
    else:
        print('Filelabel not known.')
        print('i = %s , file_labels = %s' %(i, file_labels[i]))
        
    df = pd.read_csv(file, header=0, names=column_names, parse_dates=['datum', 'tijd'])
    dfs.append(df)
```

## Creating a more easily usable datetime column
Transform the KNMI datetime formatting to something that is more easily usable (YYYY-MM-DD HH:MM:SS).

```python
# Get the 'tijd' column ready for datetime formatting
# It's not a nice way, but it works.
import datetime as dt

i = 0

for i,df in enumerate(dfs):
    df.tijd = df.tijd.str.replace(' ', '0') # change '  10' to '0010'
    df.tijd = df.tijd.apply(lambda x: (x[:2]+':'+x[2:])+':00') # change '0010' to '00:10:00'
    df.tijd = df.tijd.str.replace('24:00:00', '23:59:59')

    # tijd & datum need to be str in order to combine them
    df.tijd = pd.to_datetime(df.tijd, format='%H:%M:%S').dt.time.astype('str') 
    df.datum = df.datum.dt.strftime('%Y-%m-%d').astype('str')

    df['datetime'] = df.datum+' '+df.tijd
    df.datetime = pd.to_datetime(df.datetime, format='%Y-%m-%d %H:%M:%S')

    # Delete the no longer needed frames
    del df['datum'], df['tijd'], df['stn']

    # Set datetime as index
    df = df.set_index(['datetime'])
    
    # Resample to be able to merge
    df = df.resample('10min').mean()

    # Update the df in the dfs list
    dfs[i] = df
```

```python
"""
Need to merge the frames of the same year, then concat those together.
Ideal would be to use:

from functools import reduce
t = reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True),dfs)

But this messes things up, so lets do it manually.
"""
from functools import reduce

dfs_2015 = []
dfs_2016 = []
dfs_2017 = []
dfs_2018 = []

# Create lists of the dfs per year
for i,file in enumerate(files):
    if file_labels[i][1] == '2015':
        dfs_2015.append(dfs[i])
    elif file_labels[i][1] == '2016':
        dfs_2016.append(dfs[i])
    elif file_labels[i][1] == '2017':
        dfs_2017.append(dfs[i])
    elif file_labels[i][1] == '2018':
        dfs_2018.append(dfs[i])
    else:
        print('Filelabel not known.')
        print('i = %s , file_labels = %s' %(i, file_labels[i]))

df_2015 = reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True),dfs_2015)
df_2016 = reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True),dfs_2016)
df_2017 = reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True),dfs_2017)
df_2018 = reduce(lambda left,right: pd.merge(left, right, how='outer', left_index=True, right_index=True),dfs_2018)
```

# Inspection of the final weather DataFrame

## Memory usage

```python
"""
df_year now is the complete df of that year.
Finally concat them together into one large df.
"""
df_final = pd.concat([df_2015, df_2016, df_2017, df_2018])
df_final.info()
```

```python
"""
Reduce the memory usage, float64 precision is not required anyway.
"""

df_final= df_final.astype(dtype='float32')
df_final.info()
```

Finally, convert the UTC timezone as used by the KNMI to the GMT+1 timezone as is used in the smart meter data by OPSCHALER.

```python
df_final.index = df_final.index + pd.Timedelta(hours=2) # UTC to gmt+1
df_final.head()
```
