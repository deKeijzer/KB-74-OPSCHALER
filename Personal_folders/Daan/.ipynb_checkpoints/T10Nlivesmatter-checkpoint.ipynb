```python
import pandas as pd
import numpy as np
from bokeh.charts import Scatter, output_file, show
import matplotlib.pyplot as plt

df = pd.read_csv("/datc/opschaler/combined_dfs_gas_smart_weather/P01S02W4827_merged_gas_smart_weather.csv",header=0,delimiter="\t",parse_dates = ['datetime'])
```

```python
df.head()
```

```python
plt.title('House ID: P01S01W0373')

plt.plot(df['datetime'][:], df['eMeter'][:], '-', color='r', linewidth=1)
plt.xlabel('Date [-]')
plt.ylabel('eMeter [kWh]')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.savefig('sample2.png', dpi=1000)
plt.show()
```

```python
plt.title('House ID: P01S01W0373')

plt.plot(df['datetime'][:], df['gasMeter'][:], '-', color='r', linewidth=1)
plt.xlabel('Date [-]')
plt.ylabel('gasMeter [m3]')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
#plt.savefig('sample2.png', dpi=1000)
plt.show()
```
