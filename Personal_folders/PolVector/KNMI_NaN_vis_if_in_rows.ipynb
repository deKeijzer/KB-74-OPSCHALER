```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('//datc//opschaler//weather_data//knmi_10_min_raw_data//output//df_combined_uncleaned.csv', delimiter='\t', parse_dates=['datetime'])
```

```python
smart_nan_test = df
smart_nan_test = smart_nan_test.reset_index()
smart_nan_test = smart_nan_test.isnull()
smart_nan_test.head()
ax = sns.heatmap(smart_nan_test, cmap='gray_r', yticklabels = 10000, cbar=False)
ax.invert_yaxis()
plt.savefig('test_nans_vis_weather_whole.png')
plt.show()
```
