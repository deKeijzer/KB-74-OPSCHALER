```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df =pd.read_csv("/datc/opschaler/smartmeter_data/P01S01W0373.csv",header=0,delimiter=";",parse_dates=['Timestamp', 'gasTimestamp'])

#df['eDay'], df['eHour'] = df['Timestamp'].str.split(' ', 1).str
#df['gasDay'], df['gasHour'] = df['gasTimestamp'].str.split(' ', 1).str
df.head()
```

```python
plt.plot(df.Timestamp, df.eMeter, '-', color='r')
plt.xlabel('Date [-]')
plt.ylabel('eMeter [?]')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
```

```python
"""
This probably is the live energy usage?
"""

plt.plot(df.Timestamp, df.ePower, '-', color='r', linewidth=0.3)
plt.xlabel('Date [-]')
plt.ylabel('ePower [?]')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
```

```python
plt.plot(df.Timestamp, df.gasMeter, '-', color='r')
plt.xlabel('Date [-]')
plt.ylabel('gasMeter [?]')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
```
